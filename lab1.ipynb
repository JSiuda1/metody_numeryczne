```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 0
```

```python
x_arr = np.arange(-100, 100, 5)
x_lin = np.linspace(-100, 100, 100)
print(f"Arrange len {x_arr.size}, linspace len {x_lin.size}")
```

```python
x_dict = {"arrange": x_arr, "linspace": x_lin}
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(fig.get_size_inches() * 1.5)
for i, (x_name, x_val) in enumerate(x_dict.items()):
    y = a * np.power((x_val - b), 2)
    ax[i].plot(x_val, y, 'og')
    ax[i].set_title(x_name.capitalize() + " y = a*(x-b)^2")
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
    ax[i].grid(True)

```

```python
def fib_req(n: int) -> int:
    if n <= 1:
        return 1
    else:
        return fib_req(n - 1) + fib_req(n - 2)
```

```python
[fib_req(i) for i in range(0, 10)]
```

```python
fib_lambda = lambda n: 1 if n <= 1 else fib_lambda(n - 1) + fib_lambda(n - 2)
```

```python
[fib_lambda(i) for i in range(0, 10)]
```
